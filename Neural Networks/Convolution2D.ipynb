```python
import numpy as np
from skimage.io import imread
from skimage.color import rgb2grey
from scipy.ndimage.filters import convolve
import matplotlib.pyplot as plt
```

```python
color_image = imread('res/convolution/baboon.png', as_gray=True)
I = rgb2grey(color_image)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(I, cmap='gray')
plt.show()
```

```python
k1 = np.array([[1,0,-1],[1,0,-1],[1,0,-1]])
k2 = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])

plt.figure(1)

plt.subplot(121)
plt.imshow(k1, cmap='gray')

plt.subplot(122)
plt.imshow(k2, cmap='gray')

plt.show()
```

```python
Gk1 = convolve(I, k1)
Gk2 = convolve(I, k2)
```

```python
plt.figure(figsize=(30,15))

plt.subplot(121)
plt.imshow(Gk1, cmap='gray')

plt.subplot(122)
plt.imshow(Gk2, cmap='gray')

plt.show()
```
